```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the Brusselator equation
def Brusselator(state, t, a, b):
    x, y = state
    dx = a + x**2*y - (b+1)*x
    dy = b*x - x**2*y
    return dx, dy

# Set the initial state and parameters
state0 = [1, -1]
params = [1, 2.5]
t = np.linspace(0, 30, 1000)

# Solve the equation using odeint
state = odeint(Brusselator, state0, t, args=tuple(params))
x, y = state.T

# Plot the results
plt.plot(t, x, 'r', label='Prey')
plt.plot(t, y, 'b', label='Predator')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.show()
```

```python
# Stack the data and save to CSV file
data = np.column_stack((t, x, y))
np.savetxt("synthetic_Brusselator_data.csv", data, delimiter=",", header="t,z1,z2")
```
